```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['NAME_CONTRACT_TYPE'], df['TARGET'], test_size=0.2, random_state=7)
```

```python
df_train = pd.DataFrame()
df_train['NAME_CONTRACT_TYPE'] = x_train
df_train['TARGET'] = y_train

df_test= pd.DataFrame()
df_test['NAME_CONTRACT_TYPE'] = x_test
df_test['TARGET'] = y_test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# tfidf vectorizer
tfidf=TfidfVectorizer()
tfidf.fit(df['NAME_CONTRACT_TYPE'].values.astype('U'))
x_train_tfidf = tfidf.transform(df_train['NAME_CONTRACT_TYPE'].values.astype('U'))
x_test_tfidf = tfidf.transform(df_test['NAME_CONTRACT_TYPE'].values.astype('U'))
```

```python
df_train.to_csv('datatrain.csv', index=False)
df_test.to_csv('datatest.csv', index=False)
```

```python
print(x_train_tfidf.shape)
print(x_test_tfidf.shape)
```

```python
print(tfidf.vocabulary_)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report 
bayes = MultinomialNB()
bayes.fit(x_train_tfidf,y_train)
```

```python
nb_result = bayes.predict(x_test_tfidf)
from sklearn import metrics
model = metrics.accuracy_score(y_test, nb_result)
print("The Accuracy is",str('{:04.2f}'.format(model*100))+'%')
```
